# Cluster sampling 

```python
# Set the number of subsets to divide
n_splits = 5
# Randomly split the class_0_data into 5 subsets.
splits = []
class_0_data_copy = class_0_data.sample(frac=1, random_state=42).reset_index(drop=True)  
split_size = len(class_0_data_copy) // n_splits  
for i in range(n_splits):
    start_index = i * split_size
    if i == n_splits - 1:  
        split = class_0_data_copy.iloc[start_index:]
    else:
        split = class_0_data_copy.iloc[start_index:start_index + split_size]
    splits.append(split)
# Check the size of each subset.
for i, split in enumerate(splits):
    print(f"Split {i+1} size:", split.shape)
```
